```python
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
df=pd.read_csv('hiring.csv')
```

```python
df
```

```python
df.experience=df.experience.fillna("zero")
```

```python
df
```

```python
exp=df['experience']
exp
```

```python
%pip install word2number
```

```python
from word2number import w2n
```

```python
d=[]
for i in exp:
  d.append(w2n.word_to_num(i))
```

```python
df['experience']=d
```

```python
df
```

```python
df=df.rename(columns={'test_score(out of 10)':'test_score','interview_score(out of 10)':'interview_score'})
```

```python
mean=df['test_score'].mean()
```

```python
df['test_score']= df['test_score'].fillna(mean)
salary=df['salary($)']
```

```python
df=df.drop('salary($)',axis='columns')
df
```

```python
reg =linear_model.LinearRegression()
reg.fit(df,salary)
```

```python
reg.predict([[2,9,6]])
```

```python
reg.predict([[12,10,10]])
```
